# 人工データ回帰

## 独立変数から説明できるデータの回帰

```r
# 二変数の回帰
set.seed(0)
n <- 10000
x1 <- rnorm(n)
x2 <- rnorm(n)
e <- rnorm(n) * 0.1
b <- 10
y <- x1 * 3 + x2 * 0.5 + e + b
hist(y)
model <- lm(y~x1 + x2)
AIC(model)
model
```

```r
# データのロード
unzip("data-independent.zip")
df <- read.csv("data-independent.csv")
file.remove("data-independent.csv")

df
```

```r
y <- df[["y"]]
y_poisson <- df[["y_poisson"]]
y_gmm <- df[["y_gmm"]]


x <- df[paste("X", seq_len(1000), sep = "")]
x <- matrix(as.matrix(x), nrow(x), ncol(x))

# 10000行
n <- 10000
# 1000変数
ncol <- 1000

hist(y)
hist(y_poisson)
hist(y_gmm)
```

```r
model <- lm(y~., data = data.frame(x))
print(coefficients(model)[1:20])
AIC(model)
```

```r
library(glmnet)
lasso_model <- glmnet(x = x, y = y, family = "gaussian", alpha = 1)
plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x, y = y, family = "gaussian", alpha = 1)
plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:110])
```

```r
model <- glm(y_poisson ~ x, family = poisson(link = "log"))

print(coefficients(model)[1:20])
AIC(model)
```

```r
# GM
# https://cran.r-project.org/web/packages/sBIC/vignettes/GaussianMixtures.pdf
library(sBIC)
g_mix <- GaussianMixtures(maxNumComponents = 10, phi = 1, restarts = 100)
m <- sBIC(y_gmm, g_mix)
print(m)
print(paste("BIC max: ", which.max(m$BIC)))
print(paste("sBIC max: ", which.max(m$sBIC)))
```

```r
matplot(
    cbind(m$BIC - m$BIC[1], m$sBIC - m$sBIC[1]),
    pch = c(1, 3),
    col = "black",
    xlab = "Number of components",
    ylab = expression(BIC - BIC(M[1])),
    las = 1, xaxt = "n"
)
axis(1, at = 1:10)
legend("topleft",
    c(expression(BIC), expression(bar(sBIC)[1])),
    pch = c(1, 3),
    y.intersp = 1.2)
```

```r
postBIC <- function(BIC) {
    prob <- exp(BIC - max(BIC))
    prob / sum(prob)
}
normalizedProbs = rbind(
    "BIC" = postBIC(m$BIC),
    "sBIC1" = postBIC(m$sBIC))
barplot(
    normalizedProbs,
    beside = TRUE,
    col = c("white", "grey"),
    legend = c(expression(BIC), expression(bar(sBIC)[1])),
    xlab = "Number of components",
    ylab = "Probability",
    args.legend = list(y.intersp = 1.2),
    names.arg = 1:10
)
```

# Mclust混合分布モデル

```r
# Mclust混合分布モデル
library(mclust, quietly = TRUE)
fit = Mclust(y_gmm, G = 1:70, model = "V")
summary(fit)
plot(fit, what = "density", main = "", xlab = "")
cluster_nums <- apply(fit$z, 1, which.max)
```

```r
# y_gmmの回帰[0]
selected_index <- which(cluster_nums == 1)
x0 = x[selected_index,]
y0 = y_gmm[selected_index]
lasso_model <- glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:110])
```

```r
# y_gmmの回帰[1]
selected_index <- which(cluster_nums == 2)
x0 = x[selected_index,]
y0 = y_gmm[selected_index]
lasso_model <- glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:110])
```

```r
# y_gmmの回帰[2]
selected_index <- which(cluster_nums == 3)
x0 = x[selected_index,]
y0 = y_gmm[selected_index]
lasso_model <- glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x0, y = y0, family = "gaussian", alpha = 1)
# plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:110])
```

# GMの結果で用いてy_gmmの外れ値を判定してみる

```r
selected_index <- which(cluster_nums == 1)
hist(y_gmm[selected_index])
library(outliers)
grubbs.test(y_gmm[selected_index])
```

```r
grubbs.test(y_gmm[selected_index],type = 11)
```

```r
specs <- matrix(nrow = 3, ncol = 2)
```

```r
# ホテリング
selected_index <- which(cluster_nums == 1)
y0 <- y_gmm[selected_index]
mu <- mean(y0) # 標本平均
s2 <- mean((y0 - mu)^2) # 標本分散（不偏分散ではない）
a <- (y0 - mu)^2 / s2 # 異常度
thr <- qchisq(0.99, 1) # カイ二乗分布から99%閾値を求める
# plot(a) # 異常度をプロット
# segments(0, thr, 300, thr, col = "red", lty = 3, lwd = 3) # 閾値に線を引く
# x = (2m+-sqrt((-2mu)^2-4(mu^2-thr*s2)))/2
thr_high <- (2 * mu + sqrt((mu * -2)^2 - 4 * (mu^2 - thr * s2))) / 2
thr_low <- (2 * mu - sqrt((mu * -2)^2 - 4 * (mu^2 - thr * s2))) / 2
plot(y0)
segments(0, thr_high, 300, thr_high, col = "red", lty = 3, lwd = 3) # 閾値に線を引く
segments(0, thr_low, 300, thr_low, col = "red", lty = 2, lwd = 2) # 閾値に線を引く
print(s2)
specs[1,] <- c(thr_high, thr_low)
```
