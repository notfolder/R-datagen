# 人工データ回帰

## 独立変数から説明できるデータの回帰

```r
# 二変数の回帰
set.seed(0)
n <- 10000
x1 <- rnorm(n)
x2 <- rnorm(n)
e <- rnorm(n) * 0.1
b <- 10
y <- x1 * 3 + x2 * 0.5 + e + b
hist(y)
model <- lm(y~x1 + x2)
AIC(model)
model
```

```r
# データのロード
unzip("data-independent.zip")
df <- read.csv("data-independent.csv")
file.remove("data-independent.csv")

df
```

```r
y <- df[["y"]]
y_poisson <- df[["y_poisson"]]
y_gmm <- df[["y_gmm"]]


x <- df[paste("X", seq_len(1000), sep = "")]
x <- matrix(as.matrix(x), nrow(x), ncol(x))

hist(y)
hist(y_poisson)
hist(y_gmm)
```

```r
model <- lm(y~., data = data.frame(x))
print(coefficients(model)[1:20])
AIC(model)
```

```r
library(glmnet)
lasso_model <- glmnet(x = x, y = y, family = "gaussian", alpha = 1)
plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x, y = y, family = "gaussian", alpha = 1)
plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:110])
```

```r
model <- glm(y_poisson ~ x, family = poisson(link = "log"))

print(coefficients(model)[1:20])
AIC(model)
```

```r
lasso_model <- glmnet(x = x, y = y_poisson, family = "poisson", alpha = 1)
plot(lasso_model, xvar = "lambda", label = TRUE)

lasso_model_cv <- cv.glmnet(x = x, y = y_poisson, family = "poisson", alpha = 1)
plot(lasso_model_cv)
lasso_model_cv$lambda.min
print(predict(lasso_model_cv, type = "coef")[1:20])
```
